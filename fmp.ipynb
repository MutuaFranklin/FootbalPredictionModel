## FOOTBALL MATCH PREDICTION




```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv("footballMatchData.csv")
np.random.seed(0)
```

```python
df.head(10)
```

```python
dummymonbal=pd.read_csv("footballMatchData.csv")
```

```python
dummymonbal.query("1.7<=home_odds<1.9 & 3.9<=away_odds<4.2 & 1.4<=over<1.6 ")
```

```python
y=dummymonbal.home_score
```

```python
dummymonbal_features=["away_score","home_odds","draw_odds","away_odds","over","under"]
```

```python
x=dummymonbal[dummymonbal_features]
```

```python
x.describe()
```

```python
x.head()
```

```python
from sklearn.tree import DecisionTreeRegressor
dummymonbal_model=DecisionTreeRegressor(random_state=1)

dummymonbal_model.fit(x,y)
```

```python
print("making prediction of the following 5 teams")
print(x.head())
print("the predictions are")
print(dummymonbal_model.predict(x.head()))
```

```python
from sklearn.metrics import mean_absolute_error
predicted_home_score=dummymonbal_model.predict(x)
mean_absolute_error(y,predicted_home_score)
```

```python
from sklearn.model_selection import train_test_split
train_x,val_x,train_y,val_y=train_test_split(x,y,random_state=0)
dummymonbal_model=DecisionTreeRegressor()
dummymonbal_model.fit(train_x,train_y)
val_predictions=dummymonbal_model.predict(val_x)
print(mean_absolute_error(val_y, val_predictions))
```
